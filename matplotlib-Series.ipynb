```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series
```

```python
# 累加和
s1 = Series(np.random.randn(10)).cumsum()
s2 = Series(np.random.randn(10)).cumsum()
```

```python
s1.plot(kind='line', label='S1', title='this is a series', grid=True)
s2.plot(label='S2',grid=True)
plt.legend()
```

```python
# 通过子图会多个图表
fig, ax = plt.subplots(2, 1)
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
fig, ax = plt.subplots(2, 1)
s1.plot(ax=ax[0], label='S1', kind='bar')
s2.plot(ax=ax[1], label='S2')
plt.legend()
```
