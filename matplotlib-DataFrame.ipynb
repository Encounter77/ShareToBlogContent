```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
```

```python
# 默认按列画图
df.plot(kind='barh', stacked=True)
```

```python
df.plot(kind='area')
```

```python
# 按行画图
df.iloc[5].plot(kind='bar')
```

```python
# 循环按行画图
for i in df.index:
    df.iloc[i].plot(label=str(i))
plt.legend()
plt.show()
```

```python
# 按照某一列进行画图
df['A'].plot(kind='bar')
```

```python
# 快速按行画图 转置后操作
df.T.plot()
```
