```python
t = np.arange(0.0, 2.0, 0.1)
```

```python
s = np.sin(t*np.pi)
```

```python
plt.plot(t, s, 'b--',label='aaaa')
plt.plot(t*2, s ,'r--',label='bbbb')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.legend()
```
