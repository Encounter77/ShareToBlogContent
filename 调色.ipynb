```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

```python
def sinplot():
#     等差数列
    x = np.linspace(0, 14, 100)
#     设置图标大小
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0.75)'% (i, i))
    plt.legend()
```

```python
sinplot()
```
