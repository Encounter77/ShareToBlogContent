```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 等差间距数据
x = np.linspace(0.0, 5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x, y1, 'b--', label='sin(x*pi)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(pi*2x)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```

```python
# subplot()切换子图(行，列，位置) 模块1
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'b--', label='sin(x*pi)')
plt.ylabel('y1 value')
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'r--', label='sin(x*pi)')
plt.ylabel('y2 value')
plt.subplot(2, 2, 3)
plt.plot(x, y2, 'r*', label='sin(x*pi)')
plt.ylabel('y1 value')
plt.subplot(2, 2, 4)
plt.plot(x, y1, 'b*', label='sin(x*pi)')
plt.ylabel('y1 value')
```

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x, y1, 'b--')
ax[1][1].plot(x, y2, 'g--')
```
